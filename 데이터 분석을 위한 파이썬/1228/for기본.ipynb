```python
# 이름 뽑기


# 이름 리스트 만들기
name ='김지영,이승호,박수진,정민우,한가람,최재원,송지은,강태현,문서현,윤하린'
name_list = name.split(',')
day_list = '월, 화, 수, 목, 금'.split(',')
name_len = len(name_list)

#한 명 뽑아서 요일과 매치
#enumerate는 자동으로 0, 1, 2, 3.. 숫자값을 갖게 함
#enumerate()함수는 기본적으로 인덱스와 원소로 이루어진 튜플로 만들어줌.
#-> 인덱스와 원소를 각각 다른 변수에 할당하고 싶다면 인자 풀기(unpacking)를 해야 함
for cnt, i in enumerate(day_list): 
    #시드로 설정하니 고정값이 나옴.
    # randomNum = random.randint(1, name_len)
    # namePrn = name_list[randomNum]
    namePrn = random.choice(name_list)
    print(cnt+1, '번', i, namePrn)
```

```python
#python의 zip -> 두 개 이상의 자료를 보낼 수 있다

nameList = '가 나 다 라'.split(' ')
dayList = '목 금 토 일'.split(' ')

for name, day in zip(nameList, dayList):
    print(name, day)
```

```python
## 이중~ 다중 for 문

food1 = ['쌈밥', '치킨', '피자']
food2 = ['사과', '아이스크림', '커피']

data = []

for x in food1:
    # print('▶메뉴는', x1)
    for y in food2:
        #print('후식은', x2)
        data.append((x, y))
data
```

```python
# 기후관련 연설문
from wordcloud import WordCloud
import matplotlib.pyplot as plt


speech_textList = ['''Ladies and gentlemen,
Thank you for joining me today to address a matter of utmost urgency and global significance—the climate crisis. As we stand at the crossroads of environmental peril, it is imperative that we confront the harsh reality before us and embark on a collective journey toward sustainable solutions.
Our planet is in the grip of an unprecedented climate emergency, characterized by rising temperatures, extreme weather events, and ecological imbalances. The consequences of our past actions are now evident in the melting ice caps, rising sea levels, and the increasing frequency of natural disasters. The time for contemplation has passed; the time for resolute action is now.
The responsibility lies with each and every one of us. We must acknowledge the role human activities have played in exacerbating the crisis and recognize the profound impact it has on vulnerable communities, ecosystems, and future generations. Our planet is sending distress signals, and it is our duty to respond with a sense of urgency and responsibility.
Transitioning to a sustainable future demands innovation, collaboration, and a fundamental shift in our attitudes toward the environment. Governments, industries, and individuals must come together to implement comprehensive policies, invest in renewable energy, and embrace eco-friendly practices. We have the technological capabilities; what we need is the collective will to make a difference.
Education is key to fostering a global consciousness about the climate crisis. We must empower the younger generation with the knowledge and skills needed to tackle environmental challenges. Through education, advocacy, and community engagement, we can build a sustainable legacy that will benefit generations to come.
In conclusion, the climate crisis is not just an environmental challenge; it is a test of our humanity. It requires us to rise above our differences, prioritize the well-being of our planet, and work towards a future where nature and humanity coexist harmoniously. Let us seize this moment to be the stewards of change, ensuring a healthier, greener, and more sustainable world for all. Together, we can make a difference. Thank you.
''', 'a dkf df b dk a akd a', 'a b dkl dk a b a']

for speech_text in speech_textList:
    #워드 클라우드 생성
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(speech_text)

    #워드 클라우드 시각화
    plt.figure(figsize=(12, 8))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()
```
