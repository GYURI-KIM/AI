## 딕셔너리를 값으로 정렬하기
### operator 모듈 사용
 - itemgetter() => 기준에 따라 키와 값을 모으는 명령어

```python
import operator

mart = {'과자':1500, '아이스크림':1700, '휴지':4800, '고기':10500, '우유':2600}
print(sorted(mart.items(), key=operator.itemgetter(0))) # sorted() 에 key = 값을 주면 그걸 기준으로 정렬하는 것 같음

## .itemgetter()에 0을 입력하면 key를 기준으로, 1을 입력하면 value를 기준으로 정렬

# 값을 기준으로 정렬하되, 내림차순으로
# print(sorted(mart.items(), key=operator.itemgetter(1), reverse=True))

# 슬라이싱 해보기
print(sorted(mart.items(), key=operator.itemgetter(1), reverse=True)[:-2])
```

```python
# 정렬해도 원본 딕셔너리에는 영향을 주지 않기 때문에 바꾸고 싶다면 정렬한 걸 대입하든지, 새로운 변수 만들어서 저장하든지

tmp = sorted(mart.items(), key=operator.itemgetter(1), reverse=True)
print('정렬된 딕셔너리', dict(tmp)) # 리스트 형태로 리턴하니까 dict()로 감싸주어 다시 딕셔너리로 변환
print('원본 딕셔너리', mart)
```
