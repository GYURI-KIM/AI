```python
# 함수와 매개변수

# 인자 하나에 a를 리스트로 줄 경우
def mySum(a):
    sum = 0
    for i in a:
        sum+=i
    return sum
a = [1,2,3,4]
print(mySum(a))
```

### [미션] 함수 작성: site 여러 개, name 여러 개를 받아서 네이버, 다음, 구글 등의 QR 만들기

```python
## [미션여러 개의 사이트 QR 만들기

import qrcode

sites = ['https://www.naver.com', 'https://www.daum.net', 'https://www.google.com']
names = ['네이버', '다음', '구글']

def create_qr(site, name):
    for site, name in zip(sites, names):        # zip으로 리스트 두개 for돌리기~
        MyQR = qrcode.make(site)
        MyQR.save('c:/data/result/'+name+'_qr.png')
        img = Image.open('c:/data/result/'+name+'_qr.png')
        img.show()

create_qr(site, name)
```
