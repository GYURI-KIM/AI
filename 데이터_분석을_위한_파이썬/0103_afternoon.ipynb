```python
## 복습
def mysum(*args):
    sum = 0
    for x in args:
        sum+=x
    return sum

mysum(1,5, 7,5)
```

```python
class cal:
    def __init__(self):     # 클래스 인스턴트를 생성하는 순간 자동으로 맨 머첨 실행하는 바
        print('------------------- ')

    def test(x=2):
        print(x+5)

cal()
cal.test(8)
```
