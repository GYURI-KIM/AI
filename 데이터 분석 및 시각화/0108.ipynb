#### 히스토그램
- 그레이스케일 영상의 히스토그램 구하기

```python
import cv2
import sys
import matplotlib.pyplot as plt

src = cv2.imread('flower04.jpg', cv2.IMREAD_GRAYSCALE)

if src is None:
    print("Image load failed")
    sys.exit()

# 그레이스케일 
hist = cv2.calcHist([src], [0], None, [256], [0, 256])

cv2.imshow('src', src)
cv2.waitKey(1)

plt.plot(hist)
plt.show()
```
